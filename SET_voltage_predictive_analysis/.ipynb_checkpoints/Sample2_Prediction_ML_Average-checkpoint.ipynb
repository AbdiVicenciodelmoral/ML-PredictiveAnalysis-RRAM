```python
import pandas as pd
import numpy as np
import nbimporter
from MLModels import mlModels
from FeedForwardNetwork import NeuralNet, save_summary_as_image
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import torch
import seaborn as sns
import matplotlib.pyplot as plt

# Device configuration: check if there is a configured GPU available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Load the dataframes
setVoltageDf = pd.read_csv('process_data/setValues_KDE.csv')
setVoltageDf.set_index('Sample', inplace=True)
#display(setVoltageDf)
```

# Isolate the Fabrication Sample

```python
smp2 = setVoltageDf.loc['Sample_2']
smp2_X = smp2.drop('Set_voltage_kde',axis=1)
smp2_y = smp2['Set_voltage_kde']
display(smp2_X)
```

```python
setVoltageDf = setVoltageDf.drop(index='Sample_2')
idx = setVoltageDf.index.unique()
s_idx = 'Sample_2'
if s_idx in idx:
    raise ValueError('Sample in Training set')
setVoltageDf.index.unique()
```

```python
X = setVoltageDf.drop('Set_voltage_kde',axis=1)
y = setVoltageDf['Set_voltage_kde']
```

```python
smp2_pred_ada = smp2_X.copy()
smp2_pred_rfr = smp2_X.copy()
ada_predictions = []
rfr_predictions = []
window_size = 4

for _ in range(100):
    reg_Models = mlModels()
    reg_Models.runModels(X,y,grid_search=False)
    smp2_pred_ada['Set_voltage_kde'] = reg_Models.newSample(smp2_X,'ada')
    ada_predictions.append(smp2_pred_ada['Set_voltage_kde'].values)
    smp2_pred_rfr['Set_voltage_kde'] = reg_Models.newSample(smp2_X,'rfr')
    rfr_predictions.append(smp2_pred_rfr['Set_voltage_kde'].values)
    

def average_line(data,model_name,window_size):
    
    # Calculate the cumulative sum of values at each index
    cumulative_sum = np.sum(data, axis=0)
    
    # Calculate the average by dividing the cumulative sum by the number of arrays
    average_line = cumulative_sum / len(data)

    # Plotting the individual lines
    for line in data:
        plt.plot(line)
    
    #Calculate ROlling Average
    rolling_avg = np.zeros(len(average_line))
    for i in range(window_size, len(average_line) + 1):
        rolling_avg[i - 1] = np.mean(average_line[i - window_size:i])

    # Plotting the average line
    plt.plot(average_line, linewidth=2, color='red', label=model_name+'_Average')

    # Customize the plot
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.legend()

    # Display the plot
    plt.show()
    print("HERE")
    return rolling_avg


smp2_pred_ada['Set_voltage_kde'] = average_line(ada_predictions,"ADA",window_size)
smp2_pred_rfr['Set_voltage_kde'] = average_line(rfr_predictions,"RFR",window_size)    
```

```python
ax = plt.gca()
smp2_pred_ada.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='#4CAF50', ax=ax,label="Adaboost", linestyle=':')
smp2_pred_rfr.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='Orange', ax=ax,label="Random Forest", linestyle='-.')
smp2.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='#069AF3', ax=ax,label="True")
plt.rcParams["figure.figsize"] = (10,6.5)
plt.rcParams.update({'font.size': 14})
plt.ylabel("Density")
#plt.legend(prop={'size': 16})
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('Sample B')
plt.tight_layout()
plt.savefig('Figures/Leave_One_Out_Sample_2_ML.jpg',dpi=300)
plt.show()
```

```python
from frechetdist import frdist
s_rfr = np.array(smp2_pred_rfr['Set_voltage_kde'].values).reshape(-1,1)
s_ada = np.array(smp2_pred_ada['Set_voltage_kde'].values).reshape(-1,1)
s_true = np.array(smp2['Set_voltage_kde'].values).reshape(-1,1)


print("Frechet distance Ada:",frdist(s_ada,s_true))
print("Frechet distance RFR:",frdist(s_rfr,s_true))
```

```python
from scipy.stats import wasserstein_distance
s_true = np.array(smp2['Set_voltage_kde'].values).reshape(-1,1).ravel().tolist()
s_ada = np.array(smp2_pred_ada['Set_voltage_kde'].values).reshape(-1,1).ravel().tolist()
s_rfr = np.array(smp2_pred_rfr['Set_voltage_kde'].values).reshape(-1,1).ravel().tolist()

# Calculate Wasserstein distance
print("Wasserstein distance Ada:", wasserstein_distance(s_ada, s_true))
print("Wasserstein distance rfr:", wasserstein_distance(s_rfr, s_true))
```
