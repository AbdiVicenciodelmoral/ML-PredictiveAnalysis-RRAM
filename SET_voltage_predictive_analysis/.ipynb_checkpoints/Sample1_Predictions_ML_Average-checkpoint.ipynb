```python
import pandas as pd
import numpy as np
import nbimporter
from MLModels import mlModels
from Conv_Regression import Conv1DNet
from getKDE import getKDE
from RBFN_Model import RBFN
from FeedForwardNetwork import NeuralNet, save_summary_as_image
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import torch
import seaborn as sns
import matplotlib.pyplot as plt

# Device configuration: check if there is a configured GPU available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Load the dataframes
setVoltageDf = pd.read_csv('process_data/setValues_KDE.csv')
setVoltageDf.set_index('Sample', inplace=True)
display(setVoltageDf)
```

# Isolate the Fabrication Sample

```python
smp1 = setVoltageDf.loc['Sample_1']
smp1_X = smp1.drop('Set_voltage_kde',axis=1)
smp1_y = setVoltageDf['Set_voltage_kde']
display(smp1)
smp1.to_csv('smp1.csv')
```

```python
setVoltageDf = setVoltageDf.drop(index='Sample_1')
idx = setVoltageDf.index.unique()
s_idx = 'Sample_1'
if s_idx in idx:
    raise ValueError('Sample in Training set')
setVoltageDf.index.unique()
```

```python
X = setVoltageDf.drop('Set_voltage_kde',axis=1)
y = setVoltageDf['Set_voltage_kde']
```

```python
smp1_pred_ada = smp1_X.copy()
smp1_pred_rfr = smp1_X.copy()
ada_predictions = []
rfr_predictions = []
window_size = 4

for _ in range(100):
    reg_Models = mlModels()
    best_params = reg_Models.runModels(X,y,grid_search=False)
    smp1_pred_ada['Set_voltage_kde'] = reg_Models.newSample(smp1_X,'ada')
    ada_predictions.append(smp1_pred_ada['Set_voltage_kde'].values)
    smp1_pred_rfr['Set_voltage_kde'] = reg_Models.newSample(smp1_X,'rfr')
    rfr_predictions.append(smp1_pred_rfr['Set_voltage_kde'].values)
    if best_params:  # only attempt to print if best_params is not None
        print("Best parameters for Random Forest: ", best_params['rfr_best_params'])
        print("Best parameters for Ada Boost: ", best_params['ada_best_params'])

    

def average_line(data,model_name,window_size):
    
    # Calculate the cumulative sum of values at each index
    cumulative_sum = np.sum(data, axis=0)
    
    # Calculate the average by dividing the cumulative sum by the number of arrays
    average_line = cumulative_sum / len(data)

    # Plotting the individual lines
    for line in data:
        plt.plot(line)
    
    #Calculate ROlling Average
    rolling_avg = np.zeros(len(average_line))
    for i in range(window_size, len(average_line) + 1):
        rolling_avg[i - 1] = np.mean(average_line[i - window_size:i])

    # Plotting the average line
    plt.plot(average_line, linewidth=2, color='red', label=model_name+'_Average')

    # Customize the plot
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.legend()

    # Display the plot
    plt.show()
    plt.savefig('Figures/AVG'+model_name+'_Sample_1.jpg',dpi=300)
    return rolling_avg


smp1_pred_ada['Set_voltage_kde'] = average_line(ada_predictions,"ADA",window_size)
smp1_pred_rfr['Set_voltage_kde'] = average_line(rfr_predictions,"RFR",window_size)   
```

```python

ax = plt.gca()
smp1_pred_ada.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='#4CAF50', ax=ax,label="Adaboost", linestyle=':')
smp1_pred_rfr.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='Orange', ax=ax,label="Random Forest", linestyle='-.')
#smp1_pred_FFN.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='#C20078', ax=ax,label="FFN", linestyle='--')
smp1.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='#069AF3', ax=ax,label="True")
#smooth_smp_FFN.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='#FFD700', ax=ax,label="FFN")
plt.rcParams["figure.figsize"] = (10,6.5)
plt.rcParams.update({'font.size': 14})
plt.ylabel("Density")
#plt.legend(prop={'size': 16})
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('Sample A')
plt.tight_layout()
plt.savefig('Figures/Leave_One_Out_Sample_1_ML.jpg',dpi=300)
plt.show()
```

```python
from frechetdist import frdist
s_rfr = np.array(smp1_pred_rfr['Set_voltage_kde'].values).reshape(-1,1)
s_ada = np.array(smp1_pred_ada['Set_voltage_kde'].values).reshape(-1,1)
s_true = np.array(smp1['Set_voltage_kde'].values).reshape(-1,1)


print("Frechet distance Ada:",frdist(s_ada,s_true))
print("Frechet distance RFR:",frdist(s_rfr,s_true))
```

```python
from scipy.stats import wasserstein_distance
s_true = np.array(smp1['Set_voltage_kde'].values).reshape(-1,1).ravel().tolist()
s_ada = np.array(smp1_pred_ada['Set_voltage_kde'].values).reshape(-1,1).ravel().tolist()
s_rfr = np.array(smp1_pred_rfr['Set_voltage_kde'].values).reshape(-1,1).ravel().tolist()

# Calculate Wasserstein distance
print("Wasserstein distance Ada:", wasserstein_distance(s_ada, s_true))
print("Wasserstein distance rfr:", wasserstein_distance(s_rfr, s_true))
```
