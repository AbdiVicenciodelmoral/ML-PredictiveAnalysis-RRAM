```python
smp1_pred_ada = smp1_X.copy()
smp1_pred_ada['Set_voltage_kde'] = reg_Models.newSample(smp1_X,'ada')
#smp1_pred_ada
```

```python
smp1_pred_rfr = smp1_X.copy()
smp1_pred_rfr['Set_voltage_kde'] = reg_Models.newSample(smp1_X,'rfr')
#smp1_pred_rfr
```

```python
# Load the dataframes

"""parameters = {
    "input_size":6, 
    "hidden_size":128, 
    "output_size":1,
    "batch_size": 100,   
    "learning_rate": 0.001,
    "num_epochs": 1000         
}"""

parameters = {
    "input_size": 6,
    "hidden_size": 128,
    "output_size": 1,
    "batch_size": 50,
    'num_channels': 7,
    "learning_rate": 0.001,
    "num_epochs": 1000,
    "l2_lambda": 0.01
}

setVoltageDf =  pd.read_csv('process_data/setValues_KDE.csv')

setVoltageDf.set_index('Sample', inplace=True)
setVoltageDf = setVoltageDf.drop(index='Sample_1')
idx = setVoltageDf.index.unique()
s_idx = 'Sample_1'
if s_idx in idx:
    raise ValueError('Sample in Training set')

display(setVoltageDf)

X_train = setVoltageDf.drop('Set_voltage_kde',axis=1)
y_train = setVoltageDf['Set_voltage_kde']

# Standard Scaling (Z-score normalization)
standard_scaler = StandardScaler()
X_scaled = standard_scaler.fit_transform(X_train)


#model = NeuralNet(X_scaled,y_train,parameters).to(device)
model = Conv1DNet(X_scaled,y_train,parameters).to(device)
model.train_model()
scaled_smp1_X = standard_scaler.transform(smp1_X)


smp1_pred_FFN = model.testSample(scaled_smp1_X,smp1_X)
display(smp1_pred_FFN)
```

```python
def moving_average(data, window_size):
    ma = np.zeros(len(data))
    for i in range(window_size, len(data) + 1):
        ma[i - 1] = np.mean(data[i - window_size:i])
    return ma

# Set the window size for the moving average
window_size = 4

# Calculate the moving average of the predictions

smooth_smp_FFN = smp1_pred_FFN.copy()
smooth_smp_FFN['Set_voltage_kde'] = moving_average(smp1_pred_FFN['Set_voltage_kde'], window_size)

# Now you can use the smoothed predictions for visualization or further analysis
#display(smooth_smp_FFN)
```

```python
from frechetdist import frdist
s_rfr = np.array(smp1_pred_rfr['Set_voltage_kde'].values).reshape(-1,1)
s_ada = np.array(smp1_pred_ada['Set_voltage_kde'].values).reshape(-1,1)
s_true = np.array(smp1['Set_voltage_kde'].values).reshape(-1,1)
s_FFN = np.array(smp1_pred_FFN['Set_voltage_kde'].values).reshape(-1,1)

print("Frechet distance Ada:",frdist(s_ada,s_true))
print("Frechet distance RFR:",frdist(s_rfr,s_true))
print("Frechet distance FFN:",frdist(s_FFN,s_true))
```

```python
from scipy.stats import wasserstein_distance
s_true = np.array(smp1['Set_voltage_kde'].values).reshape(-1,1).ravel().tolist()
s_FFN = np.array(smp1_pred_FFN['Set_voltage_kde'].values).reshape(-1,1).ravel().tolist()
s_ada = np.array(smp1_pred_ada['Set_voltage_kde'].values).reshape(-1,1).ravel().tolist()
s_rfr = np.array(smp1_pred_rfr['Set_voltage_kde'].values).reshape(-1,1).ravel().tolist()

# Calculate Wasserstein distance
print("Wasserstein distance Ada:", wasserstein_distance(s_ada, s_true))
print("Wasserstein distance rfr:", wasserstein_distance(s_rfr, s_true))
print("Wasserstein distance FFN:", wasserstein_distance(s_FFN, s_true))
```
