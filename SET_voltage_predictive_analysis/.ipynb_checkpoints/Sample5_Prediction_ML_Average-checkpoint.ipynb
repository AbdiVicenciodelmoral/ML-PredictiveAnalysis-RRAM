```python
import pandas as pd
import numpy as np
import nbimporter
from MLModels import mlModels
from FeedForwardNetwork import NeuralNet, save_summary_as_image
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import torch
import seaborn as sns
import matplotlib.pyplot as plt

# Device configuration: check if there is a configured GPU available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Load the dataframes
setVoltageDf = pd.read_csv('process_data/setValues_KDE.csv')
setVoltageDf.set_index('Sample', inplace=True)
#display(setVoltageDf)
```

# Isolate the Fabrication Sample

```python
smp4 = setVoltageDf.loc['Sample_4']
smp4_X = smp4.drop('Set_voltage_kde',axis=1)
smp4_y = smp4['Set_voltage_kde']
display(smp4_X)
```

```python
setVoltageDf = setVoltageDf.drop(index='Sample_4')
idx = setVoltageDf.index.unique()
s_idx = 'Sample_4'
if s_idx in idx:
    raise ValueError('Sample in Training set')
```

```python
X = setVoltageDf.drop('Set_voltage_kde',axis=1)
y = setVoltageDf['Set_voltage_kde']
```

```python
reg_Models = mlModels()
reg_Models.runModels(X,y)
```

```python
smp4_pred_ada = smp4_X.copy()
smp4_pred_ada['Set_voltage_kde'] = reg_Models.newSample(smp4_X,'ada')
#smp4_pred_ada
```

```python
smp4_pred_rfr = smp4_X.copy()
smp4_pred_rfr['Set_voltage_kde'] = reg_Models.newSample(smp4_X,'rfr')
#smp4_pred_rfr
```

```python
# Load the dataframes

setVoltageDf =  pd.read_csv('process_data/setVoltage_KDE.csv')

setVoltageDf.set_index('Sample', inplace=True)
setVoltageDf = setVoltageDf.drop(index='Sample_1')
idx = setVoltageDf.index.unique()
s_idx = 'Sample_1'
if s_idx in idx:
    raise ValueError('Sample in Training set')
display(setVoltageDf)

X_train = setVoltageDf.drop('Set_voltage_kde',axis=1)
y_train = setVoltageDf['Set_voltage_kde']

# Standard Scaling (Z-score normalization)
standard_scaler = StandardScaler()
X_scaled = standard_scaler.fit_transform(X_train)


parameters = {
    "input_size":6, 
    "hidden_size":128, 
    "output_size":1,
    "batch_size": 100,   
    "learning_rate": 0.001,
    "num_epochs": 1000         
}

model = NeuralNet(X_scaled,y_train,parameters).to(device)
model.train_model()
scaled_smp4_X = standard_scaler.transform(smp4_X)
smp4_pred_FFN = model.testSample(scaled_smp4_X,smp4_X)
display(smp4_pred_FFN)
```

```python
save_summary_as_image(model, 'Figures/model_summary2.png', file_format='png')
```

```python
def moving_average(data, window_size):
    ma = np.zeros(len(data))
    for i in range(window_size, len(data) + 1):
        ma[i - 1] = np.mean(data[i - window_size:i])
    return ma

# Set the window size for the moving average
window_size = 4

# Calculate the moving average of the predictions

smooth_smp_FFN = smp4_pred_FFN.copy()
smooth_smp_FFN['Set_voltage_kde'] = moving_average(smp4_pred_FFN['Set_voltage_kde'], window_size)

# Now you can use the smoothed predictions for visualization or further analysis
#display(smooth_smp_FFN)
```

```python
ax = plt.gca()
smp4_pred_ada.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='#4CAF50', ax=ax,label="Adaboost")
smp4_pred_rfr.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='#C20078', ax=ax,label="Random Forest")
smp4_pred_FFN.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='#FFD700', ax=ax,label="FFN")
smp4.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='#069AF3', ax=ax,label="True")
smooth_smp_FFN.plot(kind='line',x='SET Voltage',y='Set_voltage_kde', color='Orange', ax=ax,label="FFN")
plt.rcParams["figure.figsize"] = (10,6.5)
plt.rcParams.update({'font.size': 14})
plt.ylabel("Density")
#plt.legend(prop={'size': 16})
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.savefig('Figures/Leave_One_Out_Sample_1.jpg',dpi=300)
plt.show()
```
